# Preprocessing & selecting features 

```python
import pandas as pd
import math
import numpy as np
# load the provided data

features = [
    'weekofyear',
    'reanalysis_specific_humidity_g_per_kg', 
    'reanalysis_dew_point_temp_k', 
    'station_avg_temp_c', 
    'reanalysis_max_air_temp_k',
    'station_max_temp_c',
    'reanalysis_min_air_temp_k',
    'reanalysis_air_temp_k',
    'station_min_temp_c',   
    'reanalysis_avg_temp_k',
    'reanalysis_relative_humidity_percent', 
    'reanalysis_precip_amt_kg_per_m2', 
    'reanalysis_tdtr_k',
]

def loadData(features, labels=None): 
    train_features = pd.read_csv(features, infer_datetime_format=True)
    if labels != None:
        train_labels = pd.read_csv(labels)

    # Seperate data for San Juan
    sj_train_features = train_features[train_features.city == 'sj'].copy()
    sj_train_labels = None
    if labels != None:
        sj_train_labels = train_labels[train_labels.city == 'sj'].copy()

    # Separate data for Iquitos
    iq_train_features = train_features[train_features.city == 'iq'].copy()
    iq_train_labels = None
    if labels != None:
        iq_train_labels = train_labels[train_labels.city == 'iq'].copy()
    
    if labels != None:
        return sj_train_features, sj_train_labels, iq_train_features, iq_train_labels
    else:
        return sj_train_features, iq_train_features
```

```python
#reading in data
sj_train_features, sj_train_labels, iq_train_features, iq_train_labels = loadData("dengue_features_train.csv", "dengue_labels_train.csv")

test_data = pd.read_csv("dengue_features_test.csv")
```

```python
# select columns .
sj_train_features = sj_train_features[features]
iq_train_features = iq_train_features[features]
```

```python
#fill NaN values. 
# change all NaN values to the average value in that column.
def fillNaNvalues(featuresX): 
    featuresX = featuresX.as_matrix()
    featuresX = featuresX.transpose()
    for i in range(len(featuresX)):
        for j in range(len(featuresX[i])):
            if math.isnan(featuresX[i][j]) == True:
                sumarr = [element for element in featuresX[i] if math.isnan(element) == False]
                featuresX[i][j] = np.mean(sumarr)
    featuresX = featuresX.transpose()
```

```python
fillNaNvalues(sj_train_features)
fillNaNvalues(iq_train_features)
```

# Create new feature based on the week of year 

```python
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor

#randomforestregressor
randomforest_SJ = RandomForestRegressor(n_estimators = 200)
randomforest_IQ = RandomForestRegressor(n_estimators = 200)

sjweekofyear = (sj_train_labels['weekofyear']).as_matrix()
sjweekofyear = sjweekofyear[:, None]
sjweekofyearlabels = (sj_train_labels['total_cases']).as_matrix()

iqweekofyear = (iq_train_labels['weekofyear']).as_matrix()
iqweekofyear = iqweekofyear[:, None]
iqweekofyearlabels = (iq_train_labels['total_cases']).as_matrix() 

featureSJ = randomforest_SJ.fit(sjweekofyear, sjweekofyearlabels)
featureIQ = randomforest_IQ.fit(iqweekofyear, iqweekofyearlabels)

featureSJ = randomforest_SJ.predict(sjweekofyear)
featureIQ = randomforest_IQ.predict(iqweekofyear)

plt.plot(sjweekofyear, featureSJ, '.', label="SJforest")
plt.plot(sjweekofyear, sjweekofyearlabels, 'r.', label="SJdata")
plt.title("SJ fit")
plt.show()

plt.plot(iqweekofyear, featureIQ, '.', label="IQforest")
plt.plot(iqweekofyear, iqweekofyearlabels, 'r.', label="IQdata")
plt.title("IQ fit")
plt.legend()
plt.show()

sj_train_features['weekofyear'] = featureSJ
iq_train_features['weekofyear'] = featureIQ 
```

# XGBOOST

```python
import xgboost as xgb
from sklearn.grid_search import GridSearchCV
import pandas as pd

from sklearn.metrics import make_scorer, mean_absolute_error
from sklearn.model_selection import GridSearchCV, TimeSeriesSplit


def TimeSeriesGridSearchCV(f, grid, fit_params=None, n_splits=5, verbose=0):
    scorer = make_scorer(mean_absolute_error, greater_is_better=False)
    cv     = TimeSeriesSplit(n_splits)
    return GridSearchCV(
        f, grid, scoring=scorer, fit_params=fit_params, iid=False, refit=True,
        cv=cv, verbose=verbose, return_train_score=True)
```

```python
#xgboost
cv_params = {}
ind_params = {'learning_rate': 0.1, 'n_estimators': 10, 'subsample': 0.8, 'colsample_bytree': 0.8, 
             'objective': 'reg:linear'}

#f_sj_cv = GridSearchCV(xgb.XGBRegressor(**ind_params), 
                            #cv_params, 
                             #scoring = 'neg_mean_absolute_error', cv = 5, n_jobs = -1) 
f_sj_cv = xgb.XGBRegressor(**ind_params)
f_sj_cv = TimeSeriesGridSearchCV(f_sj_cv, cv_params)
iq_params = {}

f_iq_cv = xgb.XGBRegressor(**ind_params)
f_iq_cv = TimeSeriesGridSearchCV(f_iq_cv, cv_params)
#f_iq_cv = GridSearchCV(xgb.XGBRegressor(**ind_params), 
                            #iq_params, 
                             #scoring = 'neg_mean_absolute_error', cv = 5, n_jobs = -1)



#f_sj_cv.fit(sj_train_features, sj_train_labels['total_cases'])
```
